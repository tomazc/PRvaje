##### Odgovor 5-1-1

$O(nkdi)$

$n$ - število $d$-dimenzionalnih vektorjev

$k$ - število gruč

$i$ - število iteracij do konvergence

##### Odgovor 5-1-2

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
iris = load_iris()

X = iris.data
true_labels = iris.target

model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

for true, pred in zip(true_labels, model.labels_):
    print(f'True: {true}, Prediction: {pred}, Same: {true == pred}')
```

```python
(true_labels == model.labels_).sum()
```

##### Odgovor 5-1-3

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_blobs, make_moons
n_samples=1000
```

```python
noisy_circles, _ = make_circles(n_samples=n_samples, factor=.5, noise=.05, random_state=42)
model = KMeans(n_clusters=2, random_state=42)
model.fit(noisy_circles)

plt.figure(figsize=(10, 10))
color = {0:"red", 1:"blue"}
for c, x in zip(model.labels_, noisy_circles):
    plt.plot(x[0], x[1], ".", color=color[c], markersize=10.0)
color[1]  = "cyan"
for i, c in enumerate(model.cluster_centers_):
    plt.plot(c[0], c[1], "X", color=color[i], markersize=20.0, markeredgecolor="black", markeredgewidth=2)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
noisy_moons, _ = make_moons(n_samples=n_samples, noise=.05, random_state=42)
model = KMeans(n_clusters=2, random_state=42)
model.fit(noisy_moons)

plt.figure(figsize=(10, 10))
color = {0:"red", 1:"blue"}
for c, x in zip(model.labels_, noisy_moons):
    plt.plot(x[0], x[1], ".", color=color[c], markersize=10.0)
for i, c in enumerate(model.cluster_centers_):
    plt.plot(c[0], c[1], "X", color=color[i], markersize=20.0)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
import numpy as np
blobs, _ = make_blobs(n_samples=n_samples, random_state=1)

init_centers = np.array([[-10, -5], 
                         [ 5, 0],  
                         [ 4.5, 0.0]]) 
model = KMeans(n_clusters=3, init=init_centers, random_state=35, n_init=1)
model.fit(blobs)

plt.figure(figsize=(10, 10))
color = {0:"red", 1:"blue", 2:"green"}
for c, x in zip(model.labels_, blobs):
    plt.plot(x[0], x[1], ".", color=color[c], markersize=10.0)
color[1]  = "cyan"
for i, c in enumerate(model.cluster_centers_):
    plt.plot(c[0], c[1], "X", color=color[i], markersize=20.0, markeredgecolor="black", markeredgewidth=2)
plt.show()
```
